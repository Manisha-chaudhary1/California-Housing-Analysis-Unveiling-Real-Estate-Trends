Q2. Draw an appropriate plot to see the distribution of housing_median_age and explain your observations.

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel('/content/housing+(1).xlsx')
# Histogram
plt.figure(figsize=(8, 6))# Sets the size of the plot
plt.hist(df['housing_median_age'], bins=20, color='skyblue', edgecolor='black')# Creates a histogram
plt.title('Distribution of Housing Median Age')# Sets the title
plt.xlabel('Housing Median Age')# Labels the x-axis
plt.ylabel('Frequency')# Labels the y-axis
plt.grid(axis='y')  # Adds gridlines along the y-axis

plt.show() # Displays the histogram plot

```

To check for symmetric or not from Skewness
we can know the skewness of the above plot by using :Skewed =3*(mean-median)/std()

If skewness is less than -1 or greater than 1, the distribution is highly skewed.
If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
 If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.

```python
print("Mean : ",df['housing_median_age'].mean())
print("median : ",df['housing_median_age'].median())
print("std : ",df['housing_median_age'].std())
```


The Skewness of the above plot is -0.08 which is -0.5 to -0.1.From this it is to be concluded that it is of approximately symmetrical.

Q10.Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
# Define a function
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif 11 <= total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# Create a new column 'total_bedroom_size'
df['total_bedroom_size'] = df['total_bedrooms'].apply(categorize_bedroom_size)

# Display the dataframe
print(df)
```
